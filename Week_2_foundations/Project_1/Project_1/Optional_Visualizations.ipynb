##  Bookings by Hotel Type (Grouping & Aggregation)

### Goal:
To calculate total bookings for each hotel type using grouping and aggregation
to understand which hotel category receives higher demand.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/workspaces/Data_Science_Internship/Week_2_foundations/Project_1/Project_1/Clean.csv")
df.head()
```

```python
hotel_group = df.groupby("hotel").size().reset_index(name="total_bookings")
hotel_group
hotel_group.plot(x="hotel", y="total_bookings", kind="bar")
plt.title("Total Bookings by Hotel Type")
plt.show()
```
