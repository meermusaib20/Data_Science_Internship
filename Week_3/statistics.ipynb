**Bussiness Question :**      
***Is the average sales per order greater than 200?***




***Hypothesis definition :***              
H0: μ ≤ 200   (average sales is 200 or less)         
H1: μ > 200   (average sales is greater than 200)


```python
from scipy import stats

t_stat, p = stats.ttest_1samp(target, 200)

print("t-stat:", t_stat)
print("p-value:", p)
```

**Results :**           
***Hypothesis Testing on Sales***

We tested whether the average sales per order exceed 200.

H₀: μ ≤ 200      
H₁: μ > 200

A one-sample t-test was conducted on 9,994 sales records.

Results:

t-statistic = 4.79

p-value = 1.69 × 10⁻⁶

Since the p-value is significantly less than 0.05, we reject the null hypothesis.         
Therefore, there is strong statistical evidence that the average sales per order        
are greater than 200.
