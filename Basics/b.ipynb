```python
def lar1_sec(l1):
    lar1 = float('-inf')
    sec = float('-inf')

    for num in l1:
        if num > lar1:
            sec = lar1
            lar1 = num
        elif num > sec and num != lar1:
            sec = num

    return sec, lar1
```
