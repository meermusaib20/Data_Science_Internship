```python
#lists
arr = [1,3,2,4,5,4]
print(max(arr))
print(min(arr))
print(sum(arr))
```

```python
#unique
unique = list(set(arr))
print(unique)
```

```python
#unique but without using set
unique = []
for i in arr:
  if i not in unique:
    unique.append(i)
print(unique)
```

```python
#unique using dictionary
arr1 = [1,2,2,3,4,5,6,7,7,5,6,4,3,2]
freq = {}
for i in arr1:
  if i not in freq:
    freq[i] = 1
print(list(freq.keys()))
```

```python
#functions
def Is_even_odd(n):
  if n % 2 == 0:
    print('even')
  else:
    print('odd')
x = int(input)
Is_even_odd(x)
```

```python
def fact(n):
  fact = 1
  for i in range(1,n+1):
    fact = fact * i
  return fact
x = int(input())
print(fact(x))
```

```python
def Is_prime(n):
  Is_prime = True
  if n<= 1:
   Is_prime = False
  else:
    for i in range(2,int(n**0.5)+1):
      if n%i == 0:
        Is_prime = False
        break
  return Is_prime
x = int(input())
Is_prime(x)
```

```python
nums1 = [2,7,11,15,23,5,4,33,12]
target = int(input())
def two_sum(nums):
  seen = {}
  for i , num in enumerate(nums):
    if target-num in seen:
      return [seen[target - num],i]
    seen[num] = i
two_sum(nums1)
```
